## Idade e Faixa etaria

```python
# Idade

df_sp['IDADE_ANOS'] = (
    df_sp['NU_IDADE_N']
    .astype(str)               # garante que é string
    .apply(lambda x: int(x[1:]) if x[0] == '4' else 0)
)



# Faixa etaria

bins = [-1, 0, 4, 9, 14, 19, 39, 59, 64, 69, 79, float('inf')]

labels = [
    "<1 ano",
    "1-4",
    "5-9",
    "10-14",
    "15-19",
    "20-39",
    "40-59",
    "60-64",
    "65-69",
    "70-79",
    "80+"
]

df_sp['FAIXA_ETARIA'] = pd.cut(
    df_sp['IDADE_ANOS'],
    bins=bins,
    labels=labels,
    right=True, 
    ordered=True
).astype(object).fillna("Ignorado")

```
