# Filtragem e Recodificação dos Bancos

## Pacotes necessários

```python
import pandas as pd
import os
```

## Leitura do banco

```python
# Costela
#tabela = DBF(filename="/home/costela/Documentos/Escorpiao-series/Dados_brutos/ANIMBR07.dbf", load=True, encoding='latin1')

# Notebook
df1 = pd.read_csv("/home/usuario/Documentos/Escorpiao_series/Dados_processados/df1_concatenado.csv", encoding='latin1')
```

## Filtro e seleção de colunas

```python
# Seleção das colunas de interesse
df1 = df1[['DT_SIN_PRI', 'SEM_PRI','ANO_NASC', 'NU_IDADE_N', 'CS_SEXO', 'CS_GESTANT', 'CS_RACA', 'CS_ESCOL_N',
    'ID_MN_RESI', 'ID_OCUPA_N', 'ANT_DT_ACI', 'ANT_UF', 'ANT_MUNIC_', 'ANT_LOCALI', 'ANT_TEMPO_',
       'ANT_LOCA_1', 'MCLI_LOCAL', 'CLI_DOR', 'CLI_EDEMA', 'CLI_EQUIMO',
       'CLI_NECROS', 'CLI_LOCAL_', 'CLI_LOCA_1', 'MCLI_SIST', 'CLI_NEURO',
       'CLI_HEMORR', 'CLI_VAGAIS', 'CLI_MIOLIT', 'CLI_RENAL', 'CLI_OUTR_2',
       'CLI_OUTR_3', 'CLI_TEMPO_', 'TP_ACIDENT', 'ANI_TIPO_1', 'ANI_SERPEN',
       'ANI_ARANHA', 'ANI_LAGART', 'TRA_CLASSI', 'CON_SOROTE', 'NU_AMPOLAS',
       'NU_AMPOL_1', 'NU_AMPOL_8', 'NU_AMPOL_6', 'NU_AMPOL_4', 'NU_AMPO_7',
       'NU_AMPO_5', 'NU_AMPOL_9', 'NU_AMPOL_3', 'COM_LOC', 'COM_SECUND',
       'COM_NECROS', 'COM_COMPOR', 'COM_DEFICT', 'COM_APUTAC', 'COM_SISTEM',
       'COM_RENAL', 'COM_EDEMA', 'COM_SEPTIC', 'COM_CHOQUE', 'DOENCA_TRA',
       'EVOLUCAO', 'DT_OBITO', 'DT_ENCERRA', 'DT_DIGITA']]
```

## Recodificação

```python
# Substituição dos códigos por descrições textuais

# Sexo
df1['CS_SEXO'] = df1['CS_SEXO'].replace({
    "M": "Masculino",
    "F": "Feminino",
    "I": None
})

# Gestante
df1['CS_GESTANT'] = df1['CS_GESTANT'].replace({
    1: "1 semestre",
    2: "2 semestre",
    3: "3 semestre",
    4: "Idade gestacional ign",
    5: "Nao",
    6: "Nao se aplica",
    9: None
})

# Raça
df1['CS_RACA'] = df1['CS_RACA'].replace({
    1: "Branca",
    2: "Preta",
    3: "Amarela",
    4: "Parda",
    5: "Indigena",
    9: None
})

# Escolaridade
df1['CS_ESCOL_N'] = df1['CS_ESCOL_N'].replace({
    0: "Analfabeto",
    1: "1 a 4 incompleto",
    2: "4a completa",
    3: "5 a 8 incompleto",
    4: "Fundamental completo",
    5: "Medio incompleto",
    6: "Medio completo",
    7: "Superior incompleto",
    8: "Superior completo",
    10: "Nao se aplica",
    9: None
})

# Tempo entre acidente e atendimento
df1['ANT_TEMPO_'] = df1['ANT_TEMPO_'].replace({
    1: "0-1h",
    2: "1-3h",
    3: "3-6h",
    4: "6-12h",
    5: "12 e 24h",
    6: "24 e +h",
    9: None
})

# Local do acidente
df1['ANT_LOCA_1'] = df1['ANT_LOCA_1'].replace({
    1: "Cabeca",
    2: "Braco",
    3: "Ante-Braco",
    4: "Mao",
    5: "Dedo da Mao",
    6: "Tronco",
    7: "Coxa",
    8: "Perna",
    9: "Pe",
    10: "Dedo do Pe",
    99: None
})

# Sinais e sintomas locais
df1['MCLI_LOCAL'] = df1['MCLI_LOCAL'].replace({1: "Sim", 2: "Nao", 9: None})

df1['CLI_DOR'] = df1['CLI_DOR'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_EDEMA'] = df1['CLI_EDEMA'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_EQUIMO'] = df1['CLI_EQUIMO'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_NECROS'] = df1['CLI_NECROS'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_LOCAL_'] = df1['CLI_LOCAL_'].replace({1: "Sim", 2: "Nao", 9: None})

# Sintomas sistêmicos
df1['MCLI_SIST'] = df1['MCLI_SIST'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_NEURO'] = df1['CLI_NEURO'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_HEMORR'] = df1['CLI_HEMORR'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_VAGAIS'] = df1['CLI_VAGAIS'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_MIOLIT'] = df1['CLI_MIOLIT'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_RENAL'] = df1['CLI_RENAL'].replace({1: "Sim", 2: "Nao", 9: None})
df1['CLI_OUTR_2'] = df1['CLI_OUTR_2'].replace({1: "Sim", 2: "Nao", 9: None})

# Tempo de coagulação
df1['CLI_TEMPO_'] = df1['CLI_TEMPO_'].replace({
    1: "Normal",
    2: "Alterado",
    9: "Nao realizado"
})

# Tipo de acidente
df1['TP_ACIDENT'] = df1['TP_ACIDENT'].replace({
    1: "Serpente",
    2: "Aranha",
    3: "Escorpiao",
    4: "Lagarta",
    5: "Abelha",
    6: "Outro",
    9: None
})

# Tipo de serpente
df1['ANI_SERPEN'] = df1['ANI_SERPEN'].replace({
    1: "Botropico",
    2: "Crotalico",
    3: "Elapidico",
    4: "Laquetico",
    5: "Serpente nao peconhenta",
    9: None
})

# Tipo de aranha
df1['ANI_ARANHA'] = df1['ANI_ARANHA'].replace({
    1: "Foneutrismo",
    2: "Loxoscelismo",
    3: "Latrodectismo",
    4: "Outra aranha",
    9: None
})

# Tipo de lagarta
df1['ANI_LAGART'] = df1['ANI_LAGART'].replace({
    1: "Lonomia",
    2: "Outra lagarta",
    9: None
})

# Classificação do acidente
df1['TRA_CLASSI'] = df1['TRA_CLASSI'].replace({
    1: "Leve",
    2: "Moderado",
    3: "Grave",
    9: None
})

# Soroterapia
df1['CON_SOROTE'] = df1['CON_SOROTE'].replace({1: "Sim", 2: "Nao", 9: None})

# Complicações
df1['COM_LOC'] = df1['COM_LOC'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_SECUND'] = df1['COM_SECUND'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_NECROS'] = df1['COM_NECROS'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_COMPOR'] = df1['COM_COMPOR'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_DEFICT'] = df1['COM_DEFICT'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_APUTAC'] = df1['COM_APUTAC'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_SISTEM'] = df1['COM_SISTEM'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_RENAL'] = df1['COM_RENAL'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_EDEMA'] = df1['COM_EDEMA'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_SEPTIC'] = df1['COM_SEPTIC'].replace({1: "Sim", 2: "Nao", 9: None})
df1['COM_CHOQUE'] = df1['COM_CHOQUE'].replace({1: "Sim", 2: "Nao", 9: None})

# Doença pré-existente
df1['DOENCA_TRA'] = df1['DOENCA_TRA'].replace({
    1: "Sim",
    2: "Nao",
    9: None
})

# Evolução
df1['EVOLUCAO'] = df1['EVOLUCAO'].replace({
    1: "Cura",
    2: "Obito por ap",
    3: "Obito por outra causa",
    9: None
})

```

## Idade e Faixa etaria

```python
# Idade tipo (hora, dia, mes, ano) --> primeiro digito

df1['IDADE_TIPO'] =  df1['NU_IDADE_N'].astype(str).str[0]

df1['IDADE_TIPO'] = df1['IDADE_TIPO'].replace({
    "1": "Hora",
    "2": "Dia",
    "3": "Mes",
    "4": "Ano"})


# Idade completa (horas, dias, meses, anos) --> todos os digitos exceto o primeiro
df1['IDADE_COMPLETA'] = df1['NU_IDADE_N'].astype(str).str[1:]


# Idade em anos (< 1 ano = 0)
df1['IDADE_ANOS'] = (
    df1['NU_IDADE_N']
    .astype(str)               # garante que é string
    .apply(lambda x: int(x[1:]) if x[0] == '4' else 0)
)


# Faixa etaria

bins = [-1, 0, 4, 9, 14, 19, 39, 59, 64, 69, 79, float('inf')]

labels = [
    "<1 ano",
    "1-4",
    "5-9",
    "10-14",
    "15-19",
    "20-39",
    "40-59",
    "60-64",
    "65-69",
    "70-79",
    "80+"
]

df1['FAIXA_ETARIA'] = pd.cut(
    df1['IDADE_ANOS'],
    bins=bins,
    labels=labels,
    right=True, 
    ordered=True
).astype(object).fillna("Ignorado")

```

## Semana Epidemiológica e Ano dos Sintomas

```python
# Extrair ano da data de inicio dos sintomas
df1['DT_SIN_PRI'] = pd.to_datetime(df1['DT_SIN_PRI'])  # se a data for no formato brasileiro
df1['ANO'] = df1['DT_SIN_PRI'].dt.year

# Semana de inicio dos sintomas
df1['SEM_PRI'] = df1['SEM_PRI'].astype(str).str[4:]
```
