## Idade e Faixa etaria

A variável "NU_IDADE" é composta de 4 digitos. O primeiro representa a classe/tipo de tempo relacionado a idade (hora, dia, mes, ano) e os outros três dígitos representam a idade numérica padrão. Então, precisamos decodificar em outras duas variáveis para poder controlar nas análises, sejam "IDADE_TIPO" para tipo hora, dia, mes e ano e, "IDADE_COMPLETA" para o numério tradicional, que será compreendido em conjunto com "IDADE_TIPO".
1. Definimos tipo como string e capturamos somente o primeiro dígito, que é atribuído a uma nova variável "IDADE_TIPO".  
2. Recodificamos "IDADE_TIPO", que representa o tipo de classe de idade registrada.
3. Criamos nova coluna "IDADE_COMPLETA" a partir dos três últimos dígitos de "NU_IDADE"
4. Não vamos usar idade como mês ou hora, então recodificamos tudo que for igual a isso como Zero, para fins de visualização e estatística.
5. Criamos categorias de faixas etárias a partir do formato padrão do SINAN/TabNet.


```python
# Idade tipo (hora, dia, mes, ano) --> primeiro digito

## 1. Define tipo como string e captura somente o primeiro dígito
df['IDADE_TIPO'] =  df['NU_IDADE_N'].astype(str).str[0]

## 2. Recodifica IDADE_TIPO
df['IDADE_TIPO'] = df['IDADE_TIPO'].replace({
    "1": "Hora",
    "2": "Dia",
    "3": "Mes",
    "4": "Ano"})


# 3. Cria Idade completa (horas, dias, meses, anos) --> todos os digitos exceto o primeiro
df['IDADE_COMPLETA'] = df['NU_IDADE_N'].astype(str).str[1:]


# 4. Idade em anos (< 1 ano = 0)
df['IDADE_ANOS'] = (
    df['NU_IDADE_N']
    .astype(str)               # garante que é string
    .apply(lambda x: int(x[1:]) if x[0] == '4' else 0)
)


# 5. Faixa etaria

bins = [-1, 0, 4, 9, 14, 19, 39, 59, 64, 69, 79, float('inf')]

labels = [
    "<1 ano",
    "1-4",
    "5-9",
    "10-14",
    "15-19",
    "20-39",
    "40-59",
    "60-64",
    "65-69",
    "70-79",
    "80+"
]

df['FAIXA_ETARIA'] = pd.cut(
    df['IDADE_ANOS'],
    bins=bins,
    labels=labels,
    right=True, 
    ordered=True
).astype(object).fillna("Ignorado")

```

## Ano, Mês e Semana Epidemiológica dos Sintomas

1. Definimos a coluna "DT_SIN_PRI" como date/time brasileiro.  
2. Criamos colunas para ano e mês.
3. A coluna "SEM_PRI" é composta por seis dígitos, sendo os quatro primeiros referentes ao ano da semana e os dois ultimos à semana propriamente. Capturamos os dois ultimos e atribuímos à própria variável.

```python
# Extrair ano da data de inicio dos sintomas

# 1. Define tipo data
df['DT_SIN_PRI'] = pd.to_datetime(df['DT_SIN_PRI'])  # se a data for no formato brasileiro

# 2. Cria colunas para ano e mês a partir da data
df['ANO'] = df['DT_SIN_PRI'].dt.year # ano de início dos sintomas
df['MES'] = df['DT_SIN_PRI'].dt.month # mês de início dos sintomas

# 3. Define semana de inicio dos sintomas
df['SEM_PRI'] = df['SEM_PRI'].astype(str).str[4:]
```

## Join ocupação

O objetivo principal é enriquecer o seu DataFrame principal (df). Atualmente, ele possui apenas o código numérico da ocupação (ID_OCUPA_N), o que não é muito legível para humanos. O código busca na tabela de referência (cbo) qual é o nome/descrição correspondente a esse código e adiciona essa informação ao seu conjunto de dados.

```python
# Merge com a tabela de ocupações (CBO)
df = df.merge(
    cbo[['CBO', 'OCUPACAO']],      # traz só as colunas necessárias (CBO=códigos; OCUPACAO=Descritivos dos codigos)
    left_on='ID_OCUPA_N', # coluna do dataframe original com código CBO
    right_on='CBO', # coluna da tabela de ocupações que contém os códigos
    how='left') # coluna da esquerda (dataframe original) é mantida independente de ter um correspondente na tabela de CBO
```

## Join população

O objetivo é identificar o município de residência e o tamanho da população para cada registro (linha) do conjunto de dados (df). O DataFrame original possui apenas o código do município (um número), este merge "traduz" esse código para o nome da cidade e anexa o dado populacional, permitindo análises como cálculos de incidência por 100 mil habitantes ou agrupamentos por cidade.

```python
# Merge com a tabela de população (IBGE)
df = df.merge(
    pop[['ID_MN_RESI', 'NOME_MUNI','POPULACAO']], 
    left_on='ID_MN_RESI', 
    right_on='ID_MN_RESI', 
    how='left')
```

## Join RRAS, DRS, CIR 

Insere no dataframe a macrorregião (RRAS), região adminsitrativa (DRS) e colegiado intergestores regionais (CIR) que o município está inserido.
O join está sendo feito pelo nome. Como os dados do dataframe e das regiões já foram filtrados previamente para o estado de São Paulo, não haverá problemas de homônimos (há municípios com mesmo nome em estados diferentes), do contrário, o mais prudente seria fazer o join por código de IBGE de 6 ou 7 dígitos.

```python
df = df.merge(
    reg[['NOME_MUNI','MACRO_NOME','DRS_NOME','CIR_NOME']],
    on='NOME_MUNI',
    how='left'
)
```
