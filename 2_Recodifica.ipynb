## Recodificação

```python
# Substituição dos códigos por descrições textuais

# Sexo
df['CS_SEXO'] = df['CS_SEXO'].replace({
    "M": "Masculino",
    "F": "Feminino",
    "I": None
})

# Gestante
df['CS_GESTANT'] = df['CS_GESTANT'].replace({
    "1": "1 semestre",
    "2": "2 semestre",
    "3": "3 semestre",
    "4": "Idade gestacional ign",
    "5": "Nao",
    "6": "Nao se apl    ica",
    "9": None
})

# Raça
df['CS_RACA'] = df['CS_RACA'].replace({
    "1": "Branca",
    "2": "Preta",
    "3": "Amarela",
    "4": "Parda",
    "5": "Indigena",
    "9": None
})

# Escolaridade
df['CS_ESCOL_N'] = df['CS_ESCOL_N'].replace({
    "00": "Analfabeto",
    "01": "1 a 4 incompleto",
    "02": "4a completa",
    "03": "5 a 8 incompleto",
    "04": "Fundamental completo",
    "05": "Medio incompleto",
    "06": "Medio completo",
    "07": "Superior incompleto",
    "08": "Superior completo",
    "10": "Nao se aplica",
    "09": None
})

# Tempo entre acidente e atendimento
df['ANT_TEMPO_'] = df['ANT_TEMPO_'].replace({
    "1": "0-1h",
    "2": "1-3h",
    "3": "3-6h",
    "4": "6-12h",
    "5": "12 e 24h",
    "6": "24 e +h",
    "9": None
})

# Local do acidente
df['ANT_LOCA_1'] = df['ANT_LOCA_1'].replace({
    "01": "Cabeca",
    "02": "Braco",
    "03": "Ante-Braco",
    "04": "Mao",
    "05": "Dedo da Mao",
    "06": "Tronco",
    "07": "Coxa",
    "08": "Perna",
    "09": "Pe",
    "10": "Dedo do Pe",
    "99": None
})

# Sinais e sintomas locais
df['MCLI_LOCAL'] = df['MCLI_LOCAL'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Dor local
df['CLI_DOR'] = df['CLI_DOR'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Edema local
df['CLI_EDEMA'] = df['CLI_EDEMA'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Equimose local
df['CLI_EQUIMO'] = df['CLI_EQUIMO'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Necrose local
df['CLI_NECROS'] = df['CLI_NECROS'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Outros sintomas locais
df['CLI_LOCAL_'] = df['CLI_LOCAL_'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Sintomas sistêmicos
df['MCLI_SIST'] = df['MCLI_SIST'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Sintomas neurológicos
df['CLI_NEURO'] = df['CLI_NEURO'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Hemorragia
df['CLI_HEMORR'] = df['CLI_HEMORR'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

df['CLI_VAGAIS'] = df['CLI_VAGAIS'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

df['CLI_MIOLIT'] = df['CLI_MIOLIT'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

df['CLI_RENAL'] = df['CLI_RENAL'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

df['CLI_OUTR_2'] = df['CLI_OUTR_2'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Tempo de coagulação
df['CLI_TEMPO_'] = df['CLI_TEMPO_'].replace({
    "1": "Normal",
    "2": "Alterado",
    "9": "Nao realizado"
})

# Tipo de acidente (animal causador)
df['TP_ACIDENT'] = df['TP_ACIDENT'].replace({
    "1": "Serpente",
    "2": "Aranha",
    "3": "Escorpiao",
    "4": "Lagarta",
    "5": "Abelha",
    "6": "Outro",
    "9": None
})

# Tipo de serpente
df['ANI_SERPEN'] = df['ANI_SERPEN'].replace({
    "1": "Botropico",
    "2": "Crotalico",
    "3": "Elapidico",
    "4": "Laquetico",
    "5": "Serpente nao peconhenta",
    "9": None
})

# Tipo de aranha
df['ANI_ARANHA'] = df['ANI_ARANHA'].replace({
    "1": "Foneutrismo",
    "2": "Loxoscelismo",
    "3": "Latrodectismo",
    "4": "Outra aranha",
    "9": None
})

# Tipo de lagarta
df['ANI_LAGART'] = df['ANI_LAGART'].replace({
    "1": "Lonomia",
    "2": "Outra lagarta",
    "9": None
})

# Classificação do acidente
df['TRA_CLASSI'] = df['TRA_CLASSI'].replace({
    "1": "Leve",
    "2": "Moderado",
    "3": "Grave",
    "9": None
})

# Soroterapia
df['CON_SOROTE'] = df['CON_SOROTE'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações locais
df['COM_LOC'] = df['COM_LOC'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações secundárias
df['COM_SECUND'] = df['COM_SECUND'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Necrose
df['COM_NECROS'] = df['COM_NECROS'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Compatimental
df['COM_COMPOR'] = df['COM_COMPOR'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Déficit neurológico
df['COM_DEFICT'] = df['COM_DEFICT'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Amputação
df['COM_APUTAC'] = df['COM_APUTAC'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Sistêmicas
df['COM_SISTEM'] = df['COM_SISTEM'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Insuficiência renal
df['COM_RENAL'] = df['COM_RENAL'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Edema
df['COM_EDEMA'] = df['COM_EDEMA'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Sepse
df['COM_SEPTIC'] = df['COM_SEPTIC'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Complicações - Choque
df['COM_CHOQUE'] = df['COM_CHOQUE'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Doença pré-existente
df['DOENCA_TRA'] = df['DOENCA_TRA'].replace({
    "1": "Sim",
    "2": "Nao",
    "9": None
})

# Evolução do caso
df['EVOLUCAO'] = df['EVOLUCAO'].replace({
    "1": "Cura",
    "2": "Obito por ap",
    "3": "Obito por outra causa",
    "9": None
})

## Conversão para merge
# Conversão ndo CBO para string para merge posterior
df['ID_OCUPA_N'] = df['ID_OCUPA_N'].astype(str)
cbo['CBO'] = cbo['CBO'].astype(str)

# Conversão ID_MN_RESI para string para merge posterior
df['ID_MN_RESI'] = df['ID_MN_RESI'].astype(str)
pop['ID_MN_RESI'] = pop['ID_MN_RESI'].astype(str)
reg['NOME_MUNI'] = reg['NOME_MUNI'].astype(str)

# Con

```

## Idade e Faixa etaria

```python
# Idade tipo (hora, dia, mes, ano) --> primeiro digito

df['IDADE_TIPO'] =  df['NU_IDADE_N'].astype(str).str[0]

df['IDADE_TIPO'] = df['IDADE_TIPO'].replace({
    "1": "Hora",
    "2": "Dia",
    "3": "Mes",
    "4": "Ano"})


# Idade completa (horas, dias, meses, anos) --> todos os digitos exceto o primeiro
df['IDADE_COMPLETA'] = df['NU_IDADE_N'].astype(str).str[1:]


# Idade em anos (< 1 ano = 0)
df['IDADE_ANOS'] = (
    df['NU_IDADE_N']
    .astype(str)               # garante que é string
    .apply(lambda x: int(x[1:]) if x[0] == '4' else 0)
)


# Faixa etaria

bins = [-1, 0, 4, 9, 14, 19, 39, 59, 64, 69, 79, float('inf')]

labels = [
    "<1 ano",
    "1-4",
    "5-9",
    "10-14",
    "15-19",
    "20-39",
    "40-59",
    "60-64",
    "65-69",
    "70-79",
    "80+"
]

df['FAIXA_ETARIA'] = pd.cut(
    df['IDADE_ANOS'],
    bins=bins,
    labels=labels,
    right=True, 
    ordered=True
).astype(object).fillna("Ignorado")

```

## Semana Epidemiológica e Ano dos Sintomas

```python
# Extrair ano da data de inicio dos sintomas
df['DT_SIN_PRI'] = pd.to_datetime(df['DT_SIN_PRI'])  # se a data for no formato brasileiro
df['ANO'] = df['DT_SIN_PRI'].dt.year
df['MES'] = df['DT_SIN_PRI'].dt.month

# Semana de inicio dos sintomas
df['SEM_PRI'] = df['SEM_PRI'].astype(str).str[4:]
```

## Join ocupação

```python
# Merge com a tabela de ocupações (CBO)
df = df.merge(
    cbo[['CBO', 'OCUPACAO']],      # traz só as colunas necessárias
    left_on='ID_OCUPA_N',
    right_on='CBO',
    how='left')
```

## Join população

```python
# Merge com a tabela de população (IBGE)
df = df.merge(
    pop[['ID_MN_RESI', 'NOME_MUNI','POPULACAO']], 
    left_on='ID_MN_RESI', 
    right_on='ID_MN_RESI', 
    how='left')
```

## Join RRAS, DRS, CIR 

```python
df = df.merge(
    reg[['NOME_MUNI','MACRO_NOME','DRS_NOME','CIR_NOME']],
    on='NOME_MUNI',
    how='left'
)
```
